```python

import matplotlib.pyplot as plt
import numpy as np

A=0.8
f0=1000
phi=np.pi/2
fs=44100

# t is an array starting from -0.002 to +0.002 in steps of 1/fs
t = np.arange(-0.002,0.002,1/fs)
x= A*np.cos(2*np.pi*f0*t+phi)

#plot (xvalues,yvalues)
plt.plot(t,x)

#axis([xmin,xmax,ymin,ymax])
plt.axis([-0.002,0.002,-0.8,0.8])


plt.xlabel('time')
plt.ylabel('amplitude')

plt.show()
```

```python
# Complex sinusoid using the exp() function of numpy.
import matplotlib.pyplot as plt
import numpy as np

# Totally there are 500 samples of the sinusoid to be constructed. It's stored in the array n.
N=500
n=np.arange(-N/2,N/2) # Try n=np.arange(N)

# This is the frequency part of the sinusoid
k=6

# This equation builds the sinusoid based on frequency k and n samples. It contains both real and imaginary terms.
# Try changing N and k and see different results. 
s=np.exp(1j*2*np.pi*k*n/N)

#plt.plot(n,np.real(s))
plt.plot(n,np.imag(s))
plt.axis([-N/2,N/2-1,-1,1]) #Try plt.axis([0,N-1,-1,1])
plt.xlabel('n')
plt.ylabel('amplitude')

plt.show()



```

```python
import numpy as np
import matplotlib.pyplot as plt

#This is the input signal. The input signal is a complex sinusoid. Which means the DFT should show only 1 frequency component. 
N  = 64
k0 = 7
x  = np.exp(1j*2*np.pi*k0/N*np.arange(64))

# X : this is the output array of Spectral Samples. 
X = np.array([])

for k in range(N):
    # Create complex exponential using n sample index (0 to N) using exp function
    s=np.exp(1j*2*np.pi*k/N*np.arange(N))
    X=np.append(X,sum(x*np.conjugate(s)))
    
plt.plot(np.arange(N),abs(X))
plt.axis([0, (N-1),0,N])

plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

#This is the input signal. The input signal is a cosine wave. You will see two peaks. 

N  = 64
k0 = 7 # When K is a decimal number, the DFT will have all sample values > 0. 
x  = np.cos(2*np.pi*k0/N*np.arange(64))

# X : this is the output array of Spectral Samples. 
X = np.array([])

for k in range(N):
    # Create complex exponential using n sample index (0 to N) using exp function
    s=np.exp(1j*2*np.pi*k/N*np.arange(N))
    X=np.append(X,sum(x*np.conjugate(s)))
    
plt.plot(np.arange(N),abs(X))
plt.axis([0, (N-1),0,N])

plt.show()
```

```python
#Inverse DFT 
import numpy as np
import matplotlib.pyplot as plt

#x is the signal for which we will find the DFT and also get the iDFT.
N=64
k0=7
x=np.cos(2*np.pi*k0*np.arange(N)/N)

#First, let's find the DFT of x, i.e. X
# For that, you have to multiply each sample of the input signal with each sinusoid and take a sum
# We know that there will be as many sinusoids as there are samples of the signal. We have N samples of the signal, so we have N sinusoids

X=np.array([])

nv = np.arange(-N/2,N/2) #nv is n vector
kv = np.arange(-N/2,N/2)  #kv is n vector

for k in kv:
    s = np.exp(1j*2*np.pi*k/N*nv)
    X = np.append(X,sum(x*np.conjugate(s)))
    
plt.plot(kv,abs(X))
plt.axis([-N/2,N/2-1,0,N])
plt.show()
    
y = np.array([])
for n in nv:
    s = np.exp(1j*2*np.pi*n/N*kv)
    y = np.append(y,1.0/N * sum(X*s))

print(y.shape)
plt.plot(kv,(y))
plt.axis([-N/2,N/2-1,-1,1])

plt.show()
    
```
